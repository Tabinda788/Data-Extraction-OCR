```python
zip_pattern = r'\b[A-Z]{2}\d{5}\b'
file_no_pattern = r'\b\d{4}\sOn\sFile\b'
address_pattern = r'\b\d+\s+[A-Z\d]+(?:\s+[A-Z.]+)+\b'
file_name_list, name_list, zip_list,state_list, file_no_list, adress_list = [],[],[],[],[],[]



image_folder="/content/drive/MyDrive/Images"
import re
for name in os.listdir(image_folder):
  if name.endswith(".jpg") and name.startswith("page"):
    image_path = image_folder + "/" + name
    result = ocr_model.ocr(image_path)
    parsed_words = []

    for sublist in result:
        for subsublist in sublist:
            if isinstance(subsublist[-1], tuple) and len(subsublist[-1]) == 2 and isinstance(subsublist[-1][0], str):
                word = subsublist[-1][0]
                parsed_words.append(word)
    file_name = name + '.pdf'
    first_name = [ele[5:] for ele in parsed_words if ele.startswith("XN")]
    zip = [ele[2:] for ele in parsed_words if re.findall(zip_pattern, ele)]
    state = [ele[:2] for ele in parsed_words if re.findall(zip_pattern, ele) ]
    file_no =  [ele[:4] for ele in parsed_words if re.findall(file_no_pattern, ele)]
    adress = [ele for ele in parsed_words if re.findall(address_pattern, ele) ]
    # print(parsed_words)
    file_name_list.append(name[5:-4] + '.pdf')
    name_list.append(first_name[0])
    zip_list.append(zip[0])
    state_list.append(state[0])
    try:
      file_no_list.append(file_no[0])
    except:
      file_no_list.append("")
    try:
      adress_list.append(adress[0])
    except:
      adress_list.append("")



```

```python
import pandas as pd
data = {'File Name': file_name_list, 'Full Name' : name_list, 'File No' : file_no_list, 'Address' : adress_list,
        'City' : state_list, 'ZIP' : zip_list}


df = pd.DataFrame(data)
df.to_csv("/content/buc-06-combine.csv")
```
